```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(100)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
plt.subplot(2,2,1)
plt.plot(x,-x**2,color='g')
plt.subplot(2,2,2)
plt.plot(x,-x**3,color='r')
plt.subplot(2,2,3)
plt.plot(x,-2*x,color='b')
plt.subplot(2,2,4)
plt.plot(x,-2**x,color='y')
plt.show()
```
